```python
#Visual 1 - Popularity vs Budget

import matplotlib.pyplot as plt

# Scatter plot of Popularity vs Budget
plt.figure(figsize=(10, 6))
plt.scatter(combined_data['budget'], combined_data['popularity'], alpha=0.5)
plt.title('Movie Popularity vs Budget')
plt.xlabel('Budget')
plt.ylabel('Popularity')
plt.grid(True)
plt.show()
```

## Research question: What factors influence the popularity of movies?

```python
# Convert release_date to datetime
combined_data['release_date'] = pd.to_datetime(combined_data['release_date'], errors='coerce')

# Extract the year from release_date
combined_data['release_year'] = combined_data['release_date'].dt.year

# Calculate the average popularity per year
popularity_by_year = combined_data.groupby('release_year')['popularity'].mean()

# Line plot of Popularity over Time
plt.figure(figsize=(12, 8))
plt.plot(popularity_by_year, marker='o', linestyle='-', color='b')
plt.title('Average Movie Popularity Over Time')
plt.xlabel('Year')
plt.ylabel('Average Popularity')
plt.grid(False)
plt.show()
```

** Answer to research question:**

**This line plot shows the average popularity of movies over time. By examining this plot, we can identify any trends or changes in movie popularity over the years. If we see an upward or downward trend, it would suggest that the factors influencing movie popularity have changed over time. This insight can be useful for understanding how audience preferences have evolved and for predicting future trends in movie popularity.**

```python
import pandas as pd
import ast
import matplotlib.pyplot as plt

# Load the combined data (if necessary)
combined_data = pd.read_csv('combined_cleaned_data.csv')

# Function to explode the 'genres' column and keep individual genres
def explode_genres(df):
    df = df.copy()
    df['genres'] = df['genres'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else [])
    df = df.explode('genres')
    df['genres'] = df['genres'].apply(lambda x: x['name'] if isinstance(x, dict) else x)
    return df

# Explode the 'genres' column
exploded_data = explode_genres(combined_data)

# Calculate the average popularity for each genre
genre_popularity = exploded_data.groupby('genres')['popularity'].mean().sort_values(ascending=False)

# Display the most popular genres
print(genre_popularity.head(5))

# Bar plot of the most popular genres
plt.figure(figsize=(12, 8))
genre_popularity.head(10).plot(kind='bar', color='skyblue')
plt.title('Most Popular Genres')
plt.xlabel('Genre')
plt.ylabel('Average Popularity')
plt.grid(False)
plt.show()
```
